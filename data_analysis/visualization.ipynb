VISUALIZATION:
-    LINE PLOTS - GROWTH/ DECAY
-    BAR CHARTS - INDIVIDUAL COMPARISION B/W VALUES
-    PIE CHARTS - INDIVIDUAL COMPARISION B/W VALUES BY PERCENTAGE
-    SCATTER PLOTS  - COMPARING 2 OR 3 VALUES BY THEIR GROUPING / CLUSTERS
-    HISTOGRAMS - FREQUENCY WISE DISTRIBUTION OF DATA
-    AREA PLOT  - SAME AS LINE
-    3D PLOT    - FOR 3 DIMENSIONAL DATA UNDERSTANDING

```python
t = np.arange(-np.pi * 3, np.pi*3, 0.5)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(t,x,'ro:')
plt.show()        # [<matplotlib.lines.Line2D at 0x24189b0bac0>] to remove this
```

```python
fig, ax = plt.subplots(figsize = (15,5))
plt.plot(t, y, 'gs--')
plt.plot(t, x, 'r--')
plt.show()
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.sort_values(by='age',inplace=True)
```

```python
fig,ax = plt.subplots(figsize =(15,5))
plt.plot(data['age'],'ro')
plt.show()
```

```python
fig,ax = plt.subplots(figsize =(15,5))
plt.bar(x=data.index,height=data['age'])
plt.show()
```

```python
fig,ax = plt.subplots(figsize =(15,5))
data['fare'].plot(kind= 'bar', ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots(figsize =(15,5))
data['fare'].tail(25).plot(kind= 'bar', ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots(figsize =(15,5))
data['fare'].tail(25).plot(kind= 'bar', ax=ax,color=['blue','black'])
plt.show()
```

```python
fig,ax = plt.subplots(figsize =(15,15))
data['fare'].tail(25).plot(kind= 'barh', ax=ax,color=['blue','black'])
plt.show()
```

```python
data2 = sns.load_dataset('penguins')
data2.head()
```

```python
data2['species'].value_counts()
```

```python
data2['species'].value_counts().plot(kind='bar',color =['olive','skyblue','orange'])
plt.show()
```

```python
fig,ax = plt.subplots(facecolor='yellow')
data2['species'].value_counts().plot(kind='pie',
                                    colors =['olive','skyblue','orange'],
                                    autopct= '%.1f',
                                    shadow=True
                                    )
plt.show()
```

```python
px.scatter_3d(data2, x='bill_length_mm', y='bill_depth_mm', z='flipper_length_mm',color='species')
```
